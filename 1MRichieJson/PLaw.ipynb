```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
mention_network = {}
with open('mentions.txt', 'rb') as f:
    for line in f:
        split_line = line.split()
        src = split_line[0]
        dst = split_line[1]
        if src not in mention_network:
            mention_network[src] = []
        mention_network[src].append(dst)    
```

```python
indegree_pairs = []
for k in mention_network:
    indegree_pairs.append((k, len(mention_network[k])))

indegree_pairs.sort(key=lambda x : x[1], reverse=True)
```

```python
_ = plt.scatter(range(len(indegree_pairs)), [indegree_pair[1] for indegree_pair in indegree_pairs])
```

```python
histogram = {}
for _, indegree in indegree_pairs:
    if indegree not in histogram:
        histogram[indegree] = 0 
    histogram[indegree] += 1 
    
```

```python
import math 
x = []
y = [] 
for k in histogram:
    x.append(math.log(k))
    y.append(math.log(histogram[k]))

_ = plt.scatter(x, y)
    
```
